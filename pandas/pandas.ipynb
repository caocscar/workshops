### 3. Read MS Access Database locally

To read in an Microsoft Access database on your local PC, we need additional modules:
1. `pyodbc` for connecting to databases
2. `os` for navigating the operating system.

```python
import os
import pyodbc
filename3 = r'person_2014Q1.accdb'
filepath3 = os.path.join(r'C:\Users\caoa\Box Sync\CSCAR\Workshop\pandas\2016 Fall',filename3)
connection_string = 'Driver={Microsoft Access Driver (*.mdb, *.accdb)};DBQ=' + filepath3
connection = pyodbc.connect(connection_string)
cursor = connection.cursor()
```

This establishes an ODBC connection to the database and creates a cursor associated with the database. 

Tip: You are required to pass in the filepath to the connection string and not just the filename even if it resides in the same directory as your script.

**NOTE**: `pyodbc` is not installed with anaconda automatically. You will need to run `conda install pyodbc` or `pip install pyodbc` in the command prompt. 

The following prints the available information associated with the cursor.

```python
for t in cursor.tables():
    print(t)
```

To read in the **Person** table into a pandas DataFrame, we use the `read_sql` method from `pd.io.sql`. We will send a SQL command using the ODBC connection we set up previously. This SQL command is asking for all the data from the Person table. There is also a `tail` method to look at the last 5 (or n) rows of the data.

```python
person = pd.io.sql.read_sql("SELECT * FROM Person",connection)
person.tail()
```

Note that the cursor was not needed to query a single table. It becomes more useful when you want to read in multiple tables from a database or when you don't know the table's name. We use a list comprehension to grab the name of the tables. We then create a for loop to query each table for its data and append it to a list as a tuple with the table's name. 

```python
tables = [t.table_name for t in cursor.tables(tableType='TABLE')]
DF = []
for t in tables:
    tbl = pd.io.sql.read_sql("SELECT * FROM " + t,connection)
    DF.append((t,tbl))
all_tables = dict(DF)
```

It is good practice to `close` the connection and/or cursor after you are done with it.

```python
cursor.close()
connection.close()
```

`all_tables` is a dictionary of the table name with the table itself. To access a particular table, use the following command to access dicts. 

```python
person = all_tables["Person"]
person.head()
```

### Selecting certain rows based on a conditional statement

Sometimes you want to select rows that only fit a certain criteria. 
Let's do conditional selections based on age.
Suppose we wanted to filter records to include only teenagers.

```python
idx = (person['age'] >= 13) & (person['age'] <= 19)
idx
```

This returns a pandas Series of boolean values. If we just want some descriptive statistics and count the number of people in this group, we can use the `sum` function or method because `True` = 1 and `False` = 0.

```python
a = sum(idx)
b = idx.sum()
print(a,b)
```

My personal preference for doing this sort of selection is using the `isin` method because of its versatility.

```python
idx = person['age'].isin([13,14,15,16,17,18,19])
```

Of course, this can get unwieldy if your age range is large like from 18 to 65. That's where the Python `range` function becomes handy.

```python
idx = person.age.isin(range(13,20))
```

You can also use `isin` with categorical variables making the syntax succinct and more readable.

```python
odot.County.isin(['BRO','HAM','BUT','NOB'])
```

instead of this longer version

```python
(odot.County == 'BRO') | (odot.County == 'HAM') | (odot.County == 'BUT') | (odot.County == 'HAM')
```

Finally, we want to return the rows from our selection, not just True and False.

```python
person[idx]
person[(person['age'] >= 13) & (person['age'] <= 19)]
```

If you want the inverse of your selection (i.e. grab False rows instead of True rows), use the `~` operator.

```python
person[~idx]
```

Another way to do conditional selection is using the `query` method. It automatically returns the new dataframe for you and the syntax is more akin to a mathematical expression.

```python
person.query('13 <= age <= 19')
```

### Binning Data

We can use the `cut` method to also re-categorize a continuous variable into discrete categories. A common usage of this method is to convert age to an age range.

```python
age_groups = [0, 12, 19, 24, 64, 120]
```

This defines the groups as half-open (or half-closed) intervals. Teenagers are (12,19]

```python
age_category = pd.cut(person['age'], age_groups,  include_lowest=True)
age_category
```

You can also customize the labels.

```python
age_labels = ["person who shouldn't be driving",'13-19','20-24','25-64','retired person']
age_category = pd.cut(person['age'], age_groups,  include_lowest=True, labels=age_labels)
age_category
```

### `Groupby` Method (Split-Apply-Combine Operations)

A lot of times when you have a dataframe, you want to do split-apply-combine operations.
The pandas documentation has a nice explanation of the process.
http://pandas.pydata.org/pandas-docs/stable/groupby.html

Suppose we were interested in a doing looking at the injury distribution of motorcyclists and helmet use. You can use nested `for` loops of course but the `groupby` method is easier. 

Let's apply some of what we learned before.
1. `merge` the units and person tables
2. select only the rows with motorcyclists (coded value = 9)

Note: `person["Document Number"]` is a string and `units["Document Number"]` is an integer. One needs to be converted to the other. If one variable is a float, and one variable is an integer, that will work without conversion like in the _Unit Number_ case

```python
person.dtypes.head(5)
```

```python
units.dtypes.head(5)
```

```python
import numpy as np
person["Document Number"] = person["Document Number"].astype(np.int64)
```

```python
personunit = person.merge(units, how='right', on=['Document Number','Unit Number'])
```

```python
personunit.head(5)
```

```python
motorcycles = personunit[personunit["Unit Type"] == 9]
motorcycles
```

Before we move on, let's change the numeric codes to descriptives.

```python
safety = {# Motorists
          1: 'None Used â€“ Vehicle Occupant',
          2: 'Shoulder Belt Only Used',
          3: 'Lap Belt Only Used',
          4: 'Shoulder And Lap Belt Used',
          5: 'Child Safety Seat Used',
          6: 'Helmet Used',
          7: 'Restraint Use Unknown',
          # Non-motorists
          8: 'None Used',
          9: 'helmet used',
          10: 'Protective pads used (elbows, knees, shins, etc)',
          11: 'Reflective clothing',
          12: 'Lighting',
          13: 'Other',
          99: 'Unknown',}
injury = {1: 'No injury',
          2: 'Possible',
          3: 'Non-incapacitating',
          4: 'Incapacitating',
          5: 'Fatal'}
airbag = {1: 'Not-Deployed',
          2: 'Deployed-Front',
          3: 'Deployed-Side',
          4: 'Deployed Both Front/Side',
          5: 'Not Applicable',
          9: 'Deployment Unknown',}
```

### Applying a Function to a DataFrame

#### Single Column / Row

Suppose we wanted to re-code traffic units as motorist or non-motorist based on their actions and add it as a new column. We can use the `apply` method with a `lambda` function. `apply` works on a row / column basis of a DataFrame,

```python
units['Motorist Flag'] = units['Pre-Crash Actions'].apply(lambda x: True if x<=14 else False )
units[['Pre-Crash Actions','Motorist Flag']].head(12)
```

Alternatively, you can define a function and pass that as an argument to the `apply` method. The two are equivalent. `lambda` functions are preferred if the function is short and can be succinctly written in one line.

```python
def determine_motorist(x):
    if x <= 14:
        return True
    else:
        return False
```

```python
units['Motorist Flag2'] = units['Pre-Crash Actions'].apply(determine_motorist)
units[['Pre-Crash Actions','Motorist Flag','Motorist Flag2']].head(8)
```

### Reshaping data

Occasionally, you will want your data in a certain format for input into another function. So far, our data has been presented in **wide** format. If we want it in a **long** format, we use the `melt` method.

```python
pd.options.display.max_rows = 8
df = crash[['County','Latitude','Longitude']]
county = df.groupby('County').mean()
county.reset_index(inplace=True)
county
```

```python
longformat = pd.melt(county, id_vars='County', value_vars=['Latitude','Longitude'])
longformat
```

To convert back from **wide** to **long** format, we use the `pivot` method.

```python
longformat.pivot(index='County', columns='variable', values='value')
```

Note: Duplicate entries when using the `pivot` method will cause an error. Use `pivot_table` instead for duplicate entries in a **long** format.

You can also use the `pivot_table` method to cast it back to wide format but also apply a function (defaults to the mean) to the group.

```python
pd.pivot_table(longformat, index='County', values='value', columns='variable', aggfunc=np.size)
```

More documentation on this topic can be found at  
http://pandas.pydata.org/pandas-docs/stable/reshaping.html

### Cross Tabulation Table

When using prediction or classification algorithms, you end up with a _predicted value_ column and an _actual value_ column. To examine the results, you need to produce a cross tabulation of frequencies to see how your prediction algorithms did. We use the `crosstab` method to produce the table.

```python
ct = crash.replace({'Crash Severity':{3:'No Cancer',1:'Benign',2:'Malignant'}, 
                    'Hit/Skip':{0:'No Cancer',1:'Benign',2:'Malignant'}})
ct.rename(columns={'Crash Severity':'Actual Value','Hit/Skip':'Predicted Value'}, inplace=True)
pd.crosstab(index=ct['Predicted Value'], columns=ct['Actual Value'], margins=True)
```

We can also get subtotals and/or normalize the counts with the `margins` and `normalize` keyword arguments.

```python
table = pd.crosstab(index=ct['Predicted Value'], columns=ct['Actual Value'], margins=True, normalize='all')
table
```

This is not in percentages nor looking very aesthetically pleasing. Using what we learned before.

```python
table.applymap(lambda x: "{:.1f}%".format(x*100))
```

To get the diagonal values (accurate predictions), you can use the `numpy.diagonal` function.

```python
np.diagonal(table)
```
